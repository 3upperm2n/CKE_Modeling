```python
import pandas as pd
import numpy as np
```

```python
trace_file = "tmp.csv"

col_names = ["Time(%)","Time","Calls","Avg","Min","Max","Name"]

df_trace = pd.read_csv(trace_file, names=col_names, engine='python')
```

```python
df_trace
```

```python
# us as the time unit
min_avg_us = 999999999999.0

for index, row in df_trace.iterrows():
    if row['Time(%)'] == '%':
#         print row['Avg']
        us = 1
        if row['Avg'] == 'ms':
            us = 1e3
        
    if row['Name'] == '[CUDA memcpy HtoD]':
        avg_in_us = float(row['Avg']) * us
#         print avg_in_us
        if avg_in_us < min_avg_us:
            min_avg_us = avg_in_us
#             print min_avg_us

print 'min avg runtime (in us) = ' + str(min_avg_us)
        
```
